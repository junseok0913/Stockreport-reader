```python
from openai import OpenAI

client = OpenAI(
    api_key=api_key,
    base_url="https://clovastudio.stream.ntruss.com/v1/openai"
)

extract_guide = """
당신은 **핵심 키워드를 추출**하는 역할입니다.
사용자의 질문에서 핵심 키워드를 최대 5개 추출하세요.

반드시 단계별로 생각하세요.
1. query를 단어별로 나눕니다.
2. 핵심 주제에 상응하는 우선순위로 추출합니다.
3. 기업명은 반드시 추출합니다.

당신은 사용자의 질문에 대해 관련 뉴스를 찾기 위해 **핵심 키워드만을 추출**하는 역할을 합니다.
질문 유형을 다음 두 가지로 나누어 키워드를 추출합니다.

1. 개념 설명형 질문
- "~가 뭐야?", "~라는게 뭔데?" "~에 대해 설명해줘"와 같은 개념 설명을 요구하는 경우
- query에서 나온 명사구 - 고유명사, 제품명, 개념을 우선순위로 키워드 추출

2. 근거 추론형 질문
- "왜 그렇게 판단했어?", "그렇게 보는 이유는 뭐야?" 와 같은 근러를 추론해야 하는 경우
- query와 context에서 중요한 키워드 - 전략, 기술, 이슈 등을 우선순위로 키워드 추출

출력은 반드시 쉼표(,)로 구분된 키워드 나열만 해야 합니다.

"""

def keyword_extract(title, context, query):
    client = OpenAI(
        api_key=api_key,
        base_url="https://clovastudio.stream.ntruss.com/v1/openai")

    structured_input = f"""
        다음은 사용자의 질문과 관련된 배경 정보입니다.

        - 제목(title): {title}
        - 문맥(context): {context.strip()}
        - 질문(query): {query}
        
        위 문맥과 질문을 바탕으로, 키워드 최대 5개를 추출해주세요"""

    response = client.chat.completions.create(
        model="HCX-005",
        messages=[
            {"role": "system", "content": extract_guide},
            {"role": "user", "content": structured_input}
            ])
    
    keyword = response.choices[0].message.content
    
    return keyword

def select_keyword(keyword, k=2):
    keywords = [kw.strip() for kw in keyword.split(",")][:k]
    return keywords
```

### get_news

```python
from dotenv import load_dotenv
import os

load_dotenv()

NAVER_CLIENT_ID = os.getenv("NAVER_CLIENT_ID")
NAVER_CLIENT_SECRET = os.getenv("NAVER_CLIENT_SECRET")
```

```python
from urllib.parse import quote
import requests
import json
import time
import os

def search_naver_news(keyword, max_count=50):
    headers = {
        "X-Naver-Client-Id": NAVER_CLIENT_ID,
        "X-Naver-Client-Secret": NAVER_CLIENT_SECRET
    }

    results = []
    start = 1
    while len(results) < max_count and start < 1000:
        url = f"https://openapi.naver.com/v1/search/news.json?query={quote(keyword)}&display=100&start={start}&sort=sim"
        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            print(f"❌ Error for '{keyword}' at start={start}: {response.status_code}")
            break

        items = response.json().get("items", [])
        for item in items:
            if "news.naver.com" not in item["link"]:
                continue  # 네이버 뉴스만 필터링
            results.append({
                "title": item["title"],
                "link": item["link"],
                "description": item["description"],
                "pubDate": item["pubDate"]
            })
            if len(results) >= max_count:
                break

        start += 100
        time.sleep(0.2)  # API rate limit 고려

    return results[:max_count]

def ensure_dir(path):
    directory = os.path.dirname(path)
    if not os.path.exists(directory):
        os.makedirs(directory)

# 키워드를 공백으로 조합하여 단일 검색
def get_news(keywords, output_path):
    os.makedirs(os.path.dirname(output_path), exist_ok=True)

    combined_query = " ".join(keywords)
    print(f"🔍 Searching for combined keywords: {combined_query}")

    news_items = search_naver_news(combined_query)

    all_results = {"query": combined_query, "results": news_items}

    ensure_dir(output_path)
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(all_results, f, ensure_ascii=False, indent=2)

    print(f"네이버 뉴스 기사 {len(news_items)}개 저장 완료 → {output_path}")
```

```python
from dotenv import load_dotenv
from urllib.parse import quote
import requests
import json
import time
import os

load_dotenv()

NAVER_CLIENT_ID = os.getenv("NAVER_CLIENT_ID")
NAVER_CLIENT_SECRET = os.getenv("NAVER_CLIENT_SECRET")

def search_naver_news(keyword, max_count=50):
    headers = {
        "X-Naver-Client-Id": NAVER_CLIENT_ID,
        "X-Naver-Client-Secret": NAVER_CLIENT_SECRET
    }

    results = []
    start = 1
    while len(results) < max_count and start < 1000:
        url = f"https://openapi.naver.com/v1/search/news.json?query={quote(keyword)}&display=100&start={start}&sort=sim"
        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            print(f"❌ Error for '{keyword}' at start={start}: {response.status_code}")
            break

        items = response.json().get("items", [])
        for item in items:
            if "news.naver.com" not in item["link"]:
                continue  # 네이버 뉴스만 필터링
            results.append({
                "title": item["title"],
                "link": item["link"],
                "description": item["description"],
                "pubDate": item["pubDate"]
            })
            if len(results) >= max_count:
                break

        start += 100
        time.sleep(0.2)  # API rate limit 고려

    return results[:max_count]

def ensure_dir(path):
    directory = os.path.dirname(path)
    if not os.path.exists(directory):
        os.makedirs(directory)

# 키워드를 공백으로 조합하여 단일 검색
def get_news(keywords, output_path):
    os.makedirs(os.path.dirname(output_path), exist_ok=True)

    combined_query = " ".join(keywords)
    print(f"🔍 Searching for combined keywords: {combined_query}")

    news_items = search_naver_news(combined_query)

    all_results = {"query": combined_query, "results": news_items}

    ensure_dir(output_path)
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(all_results, f, ensure_ascii=False, indent=2)

    print(f"네이버 뉴스 기사 {len(news_items)}개 저장 완료 → {output_path}")
```

```python
import json
import math

intention_guide = """
당신의 역할은 사용자가 문서를 읽고 던진 질문의 의도를 파악해, 간결하고 명확한 **한 줄 문장**으로 요약하는 것입니다.

다음 사항을 반드시 지켜주세요:

1. 질문 자체의 표현을 중점적으로 분석하세요. 문맥(context)은 보조적 참고로만 활용하세요.
2. 출력은 반드시 자연어 문장 형태로 하며, 예: "사용자의 질문 의도는 ~이다." 또는 "질문은 ~을 알고자 하는 목적이다." 형식으로 작성하세요.
3. 질문에 명확한 대상(제품, 기술, 기업, 전략 등)이 있다면 그것을 문장 안에 언급하세요.
4. 설명을 늘이지 마세요. 한 줄로 명확하게 의도를 요약하세요.
5. ~~를 요청하는 것이 목적이다 / 이해하려는 것이다 / 확인하려는 것이다 등으로 끝맺으세요.

반드시 한 문장으로 요약하세요.

"""

def analyze_intention(title, context, query):
    client = OpenAI(
        api_key=api_key,
        base_url="https://clovastudio.stream.ntruss.com/v1/openai")

    structured_input = f"""
        다음은 사용자의 질문과 관련된 배경 정보입니다.

        - 제목(title): {title}
        - 문맥(context): {context.strip()}
        - 질문(query): {query}
        
        위 문맥과 질문을 바탕으로, 질문의 의도를 한줄로 요약해 출력하세요"""

    response = client.chat.completions.create(
        model="HCX-005",
        messages=[
            {"role": "system", "content": intention_guide},
            {"role": "user", "content": structured_input}
            ])
    
    intention_message = response.choices[0].message.content
    
    return intention_message

skimming_guide = """
당신은 AI 어시스턴트로서, 뉴스 기사 리스트 속 각각의 뉴스가 질문 의도에 적합한지 판단하는 역할입니다.

---

[작업 지시]  
1. 뉴스 리스트 중에서, 질문의 의도와 관련성이 높은 뉴스를 판별합니다.  
   - 뉴스의 제목(title)과 간략설명(description)을 참조하세요.  
   - 너무 광범위하거나 관련 없는 뉴스는 제외합니다.  
2. 관련성이 높으면 1, 관련도가 낮으면 0을 출력합니다.
   - 예) 뉴스 1, 뉴스 2, 뉴스 3  -> 출력 = [0, 1, 0]
3. 각 뉴스의 관련도 여부만 출력합니다. 설명은 제외합니다.

---

[출력 예시]  

[1, 0, 0, 1, 1]

"""


def skimming_news(intention, news_list):
    client = OpenAI(
        api_key=api_key,
        base_url="https://clovastudio.stream.ntruss.com/v1/openai")

    structured_input = f"""
        다음은 사용자의 질문 의도와 수집된 뉴스 리스트 입니다.

        - 질문 의도(intention): {intention}
        - 뉴스 리스트(news_list): {news_list}
        
        위 문맥과 질문을 바탕으로, 필터링된 뉴스를 출력하세요"""

    response = client.chat.completions.create(
        model="HCX-005",
        messages=[
            {"role": "system", "content": skimming_guide},
            {"role": "user", "content": structured_input}
            ])
    
    filtered_news = response.choices[0].message.content
    
    return filtered_news

def chunking_news_list(intention, input_path, output_path=None): 
    with open(input_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    news_items = data.get("results", [])
    chunk_size = 5
    total = len(news_items)
    num_chunks = math.ceil(total / chunk_size)

    for i in range(num_chunks):
        chunk = news_items[i * chunk_size : (i + 1) * chunk_size]

        # LLM에 보낼 간단한 구조로 변환
        simplified_chunk = [
            {"title": item["title"], "description": item["description"]}
            for item in chunk
        ]

        result_str = skimming_news(intention, simplified_chunk)

        # 문자열 -> 리스트 변환
        try:
            decisions = json.loads(result_str)
        except:
            print(f"[WARNING] LLM 출력 파싱 실패: {result_str}")
            decisions = [0] * len(chunk)

        for j, selected in enumerate(decisions):
            chunk[j]["selected"] = selected

    if output_path:
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(data, f, ensure_ascii=False, indent=2)

    return data["results"]
```

```python
import json
import requests
from bs4 import BeautifulSoup

def crawling_news(articles, output_path):
    collected = []

    for i, article in enumerate(articles):
        link = article.get('link') or article.get('링크')
        title = article.get('title') or article.get('제목', '')

        if not link:
            print(f"[{i}] 링크 없음, 건너뜀")
            continue

        try:
            headers = {"User-Agent": "Mozilla/5.0"}
            res = requests.get(link, headers=headers, timeout=5)
            soup = BeautifulSoup(res.text, "html.parser")
            content = soup.select_one("#dic_area")
            content_text = content.get_text(separator="\n", strip=True) if content else "[본문 없음]"
        except Exception as e:
            print(f"[{i}] 본문 크롤링 실패: {e}")
            content_text = "[본문 크롤링 실패]"

        collected.append({
            "제목": title,
            "링크": link,
            "본문": content_text
        })
        print(f"[{i}] {title} - 본문 크롤링 완료")

    # 루프 밖에서 한 번만 저장
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(collected, f, ensure_ascii=False, indent=2)

    print(f"\n📝 크롤링 완료! 결과 저장: {output_path}")
```

```python
from dotenv import load_dotenv
from openai import OpenAI
import os

load_dotenv()

api_key = os.getenv("API_KEY")

client = OpenAI(
    api_key=api_key,
    base_url="https://clovastudio.stream.ntruss.com/v1/openai"
)

extract_guide = """
당신은 **핵심 키워드를 추출**하는 역할입니다.
사용자의 질문에서 핵심 키워드를 최대 5개 추출하세요.

반드시 단계별로 생각하세요.
1. query를 단어별로 나눕니다.
2. 핵심 주제에 상응하는 우선순위로 추출합니다.
3. 기업명은 반드시 추출합니다.

당신은 사용자의 질문에 대해 관련 뉴스를 찾기 위해 **핵심 키워드만을 추출**하는 역할을 합니다.
질문 유형을 다음 두 가지로 나누어 키워드를 추출합니다.

1. 개념 설명형 질문
- "~가 뭐야?", "~라는게 뭔데?" "~에 대해 설명해줘"와 같은 개념 설명을 요구하는 경우
- query에서 나온 명사구 - 고유명사, 제품명, 개념을 우선순위로 키워드 추출

2. 근거 추론형 질문
- "왜 그렇게 판단했어?", "그렇게 보는 이유는 뭐야?" 와 같은 근러를 추론해야 하는 경우
- query와 context에서 중요한 키워드 - 전략, 기술, 이슈 등을 우선순위로 키워드 추출

출력은 반드시 쉼표(,)로 구분된 키워드 나열만 해야 합니다.

"""

def keyword_extract(title, context, query):
    client = OpenAI(
        api_key=api_key,
        base_url="https://clovastudio.stream.ntruss.com/v1/openai")

    structured_input = f"""
        다음은 사용자의 질문과 관련된 배경 정보입니다.

        - 제목(title): {title}
        - 문맥(context): {context.strip()}
        - 질문(query): {query}
        
        위 문맥과 질문을 바탕으로, 키워드 최대 5개를 추출해주세요"""

    response = client.chat.completions.create(
        model="HCX-005",
        messages=[
            {"role": "system", "content": extract_guide},
            {"role": "user", "content": structured_input}
            ])
    
    keyword = response.choices[0].message.content
    
    return keyword

def select_keyword(keyword, k=2):
    keywords = [kw.strip() for kw in keyword.split(",")][:k]
    return keywords
```

```python
# 실행
title="일라이 릴리"
context="""공보험 커버에 중요한 적응증 확대 관련, 수면무호흡증(OSA)는 FDA 허가 검토를 받고 있고, 심부전(HFpEF)에 대해 FDA 허가 신청 예정이다.
노보 위고비와 H2H 임상인 SURMOUNT-5에 대한 임상3상 데이터는 올해 안에 공개될 예정이다."""
query="적응증 확대가 뭐야?"

keywords = keyword_extract(title, context, query)
final_keywords = select_keyword(keywords)
```

```python
from dotenv import load_dotenv
from urllib.parse import quote
import requests
import json
import time
import os

def search_naver_news(keyword, max_count=50):
    headers = {
        "X-Naver-Client-Id": NAVER_CLIENT_ID,
        "X-Naver-Client-Secret": NAVER_CLIENT_SECRET
    }

    results = []
    start = 1
    while len(results) < max_count and start < 1000:
        url = f"https://openapi.naver.com/v1/search/news.json?query={quote(keyword)}&display=100&start={start}&sort=sim"
        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            print(f"❌ Error for '{keyword}' at start={start}: {response.status_code}")
            break

        items = response.json().get("items", [])
        for item in items:
            if "news.naver.com" not in item["link"]:
                continue  # 네이버 뉴스만 필터링
            results.append({
                "title": item["title"],
                "link": item["link"],
                "description": item["description"],
                "pubDate": item["pubDate"]
            })
            if len(results) >= max_count:
                break

        start += 100
        time.sleep(0.2)  # API rate limit 고려

    return results[:max_count]

def ensure_dir(path):
    directory = os.path.dirname(path)
    if not os.path.exists(directory):
        os.makedirs(directory)

# 키워드를 공백으로 조합하여 단일 검색
def get_news(keywords, output_path):
    os.makedirs(os.path.dirname(output_path), exist_ok=True)

    combined_query = " ".join(keywords)
    print(f"🔍 Searching for combined keywords: {combined_query}")

    news_items = search_naver_news(combined_query)

    all_results = {"query": combined_query, "results": news_items}

    ensure_dir(output_path)
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(all_results, f, ensure_ascii=False, indent=2)

    print(f"네이버 뉴스 기사 {len(news_items)}개 저장 완료 → {output_path}")
```

```python
import json
import math

intention_guide = """
당신의 역할은 사용자가 문서를 읽고 던진 질문의 의도를 파악해, 간결하고 명확한 **한 줄 문장**으로 요약하는 것입니다.

다음 사항을 반드시 지켜주세요:

1. 질문 자체의 표현을 중점적으로 분석하세요. 문맥(context)은 보조적 참고로만 활용하세요.
2. 출력은 반드시 자연어 문장 형태로 하며, 예: "사용자의 질문 의도는 ~이다." 또는 "질문은 ~을 알고자 하는 목적이다." 형식으로 작성하세요.
3. 질문에 명확한 대상(제품, 기술, 기업, 전략 등)이 있다면 그것을 문장 안에 언급하세요.
4. 설명을 늘이지 마세요. 한 줄로 명확하게 의도를 요약하세요.
5. ~~를 요청하는 것이 목적이다 / 이해하려는 것이다 / 확인하려는 것이다 등으로 끝맺으세요.

반드시 한 문장으로 요약하세요.

"""

def analyze_intention(title, context, query):
    client = OpenAI(
        api_key=api_key,
        base_url="https://clovastudio.stream.ntruss.com/v1/openai")

    structured_input = f"""
        다음은 사용자의 질문과 관련된 배경 정보입니다.

        - 제목(title): {title}
        - 문맥(context): {context.strip()}
        - 질문(query): {query}
        
        위 문맥과 질문을 바탕으로, 질문의 의도를 한줄로 요약해 출력하세요"""

    response = client.chat.completions.create(
        model="HCX-005",
        messages=[
            {"role": "system", "content": intention_guide},
            {"role": "user", "content": structured_input}
            ])
    
    intention_message = response.choices[0].message.content
    
    return intention_message

skimming_guide = """
당신은 AI 어시스턴트로서, 뉴스 기사 리스트 속 각각의 뉴스가 질문 의도에 적합한지 판단하는 역할입니다.

---

[작업 지시]  
1. 뉴스 리스트 중에서, 질문의 의도와 관련성이 높은 뉴스를 판별합니다.  
   - 뉴스의 제목(title)과 간략설명(description)을 참조하세요.  
   - 너무 광범위하거나 관련 없는 뉴스는 제외합니다.  
2. 관련성이 높으면 1, 관련도가 낮으면 0을 출력합니다.
   - 예) 뉴스 1, 뉴스 2, 뉴스 3  -> 출력 = [0, 1, 0]
3. 각 뉴스의 관련도 여부만 출력합니다. 설명은 제외합니다.

---

[출력 예시]  

[1, 0, 0, 1, 1]

"""


def skimming_news(intention, news_list):
    client = OpenAI(
        api_key=api_key,
        base_url="https://clovastudio.stream.ntruss.com/v1/openai")

    structured_input = f"""
        다음은 사용자의 질문 의도와 수집된 뉴스 리스트 입니다.

        - 질문 의도(intention): {intention}
        - 뉴스 리스트(news_list): {news_list}
        
        위 문맥과 질문을 바탕으로, 필터링된 뉴스를 출력하세요"""

    response = client.chat.completions.create(
        model="HCX-005",
        messages=[
            {"role": "system", "content": skimming_guide},
            {"role": "user", "content": structured_input}
            ])
    
    filtered_news = response.choices[0].message.content
    
    return filtered_news

def chunking_news_list(intention, input_path, output_path=None): 
    with open(input_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    news_items = data.get("results", [])
    chunk_size = 5
    total = len(news_items)
    num_chunks = math.ceil(total / chunk_size)

    for i in range(num_chunks):
        chunk = news_items[i * chunk_size : (i + 1) * chunk_size]

        # LLM에 보낼 간단한 구조로 변환
        simplified_chunk = [
            {"title": item["title"], "description": item["description"]}
            for item in chunk
        ]

        result_str = skimming_news(intention, simplified_chunk)

        # 문자열 -> 리스트 변환
        try:
            decisions = json.loads(result_str)
        except:
            print(f"[WARNING] LLM 출력 파싱 실패: {result_str}")
            decisions = [0] * len(chunk)

        for j, selected in enumerate(decisions):
            chunk[j]["selected"] = selected

    if output_path:
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(data, f, ensure_ascii=False, indent=2)

    return data["results"]
```

```python
import json
import requests
from bs4 import BeautifulSoup

def crawling_news(articles, output_path):
    selected_news = [item for item in articles if item.get("selected") == 1]

    collected = []

    for i, article in enumerate(selected_news):
        link = article.get('link') or selected_news.get('링크')
        title = article.get('title') or selected_news.get('제목', '')

        if not link:
            print(f"[{i}] 링크 없음, 건너뜀")
            continue

        try:
            headers = {"User-Agent": "Mozilla/5.0"}
            res = requests.get(link, headers=headers, timeout=5)
            soup = BeautifulSoup(res.text, "html.parser")
            content = soup.select_one("#dic_area")
            content_text = content.get_text(separator="\n", strip=True) if content else "[본문 없음]"
        except Exception as e:
            print(f"[{i}] 본문 크롤링 실패: {e}")
            content_text = "[본문 크롤링 실패]"

        collected.append({
            "제목": title,
            "링크": link,
            "본문": content_text
        })
        print(f"[{i}] {title} - 본문 크롤링 완료")

    # 루프 밖에서 한 번만 저장
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(collected, f, ensure_ascii=False, indent=2)

    print(f"\n📝 크롤링 완료! 결과 저장: {output_path}")
```
